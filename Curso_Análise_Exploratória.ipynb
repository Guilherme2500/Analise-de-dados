# AnÃ¡lise de dados de filmes


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
filmes = pd.read_csv("tmdb_5000_movies.csv")
filmes.head()
```

* Comparando os filmes com seus votos

```python
filmes[["original_title", "vote_average"]].sort_values("vote_average", ascending=False).head(10)
```

* Mostrando quais linguas aparecem mais


```python
sns.catplot(x="original_language", kind="count", data=filmes)
```

# Contagem de linguas

```python
contagem_linguas = filmes["original_language"].value_counts().to_frame().reset_index()
contagem_linguas.columns = ["original language", "total"]
contagem_linguas
```

```python
plt.pie(contagem_linguas["total"], labels=contagem_linguas["original language"])
```

* Grafico nÃ£o recomendado para este tipo de anÃ¡lise





#Criando novo dataframe

```python
total = filmes["original_language"].value_counts()
total_ingles = total.loc["en"]
total_resto = total.sum() - total_ingles

new_data ={
    'lingua':['ingles', 'outros'],
    'total':[total_ingles, total_resto]
}

new_data = pd.DataFrame(new_data)
new_data

sns.barplot(x="lingua", y="total", data=new_data)
```

# AnÃ¡lise dos filmes que nÃ£o sejam feitos em Ingles

```python
contagem_filmes_sem_ingles = filmes.query("original_language != 'en'").original_language.value_counts()

contagem_filmes_sem_ingles.columns = ["original language", "total"]
contagem_filmes_sem_ingles
```

```python
filmes_sem_ingles = filmes.query("original_language != 'en'")
sns.catplot(x ='original_language', kind='count',data = filmes_sem_ingles,aspect=2, order = contagem_filmes_sem_ingles.index,
            palette = 'GnBu_d')
```

# Tirando a mÃ©dia e comparando dois filmes atravÃ©s da dispersÃ£o dos dados




```python
movies.head()
```

```python
rate.head()
```

```python

rate_toy_story = rate.query("movieId == 1")
rate_Jumanji = rate.query("movieId == 2")
print(len(rate_toy_story))
print(len(rate_Jumanji))
```

```python
print('MÃ©dia das notas de Toy Story %.2f' % rate_toy_story.rating.mean())
```

```python
print('MÃ©dia das notas de Jumanji %.2f' % rate_Jumanji.rating.mean())
```

* A media nÃ£o Ã© o melhor indice para comparar a disposiÃ§Ã£o de notas ou de um grande nÃºmero de dados, sendo necessÃ¡rio um boxplot.

```python
sns.boxplot([rate_toy_story.rating])
```

```python
sns.boxplot([rate_Jumanji.rating])
```

* Percebe-se que em Toy Story, ocorreu um maior nÃºmero de notas no intervalo de 2-4, sendo a maior ocorrencia em 4. JÃ¡ em Jumanji, a maior ocorrencia acontece em 3,5, sendo que existiram notas no intervalor de 1,5. Isso mostra que Toy Story Ã© um filmes melhor avaliado do que Jumanji

```python
print(rate_toy_story.rating.std())
print(rate_Jumanji.rating.std())
```

* O desvÃ£o padrÃ£o tambÃ©m Ã© uma forma de reconhecer a distribuiÃ§Ã£o daquelas notas, o que mostra se um filme estÃ¡ sendo bem avaliado ou nÃ£o. Nesse caso, o desvio dos dois filmes Ã© bem parecido, o que mostra uma disputa bem acirrada, apenas sendo capaz de visualizar atravÃ©s do grÃ¡fico.

```python
sns.scatterplot(x='userId', y="rating", data=rate_toy_story)
```

```python
sns.scatterplot(x='userId', y="rating", data=rate_Jumanji)
```
